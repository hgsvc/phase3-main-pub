```python
directory = '/home/mark/Desktop/HLA/libraries/chr6Genes/'
ExonDirectory ={}
for folder in os.listdir(directory):
    genename = str(folder.split("_")[0])
    f = str(directory)+str(folder)
    count=0
    with open(f,'r') as file:
        exonflag="FALSE"
        for line in file:
            if '>' in line:
                if 'EXON_'+str(genename) in str(line):
                    exonflag = 'TRUE'
                    count+=1
                else:
                    count+=1
                #print(line)
            else:
                continue
    file.close()
        
    ExonDirectory[folder.split("_library")[0]]={'Count':count, 'ExonFlag':exonflag}
ExonDirectory
```

```python
def ranges(nums):
    nums = sorted(nums)
    #print(nums)
    gaps = [int(s)+1 for s, e in zip(nums, nums[1:]) if s+1 < e]
    #print(gaps)
    groups = []
    tempList=[]
    for x in nums:
        
        if len(tempList)<1:
            tempList.append(x)
        else:
            if x+1 in gaps:
                tempList.append(x)
                groups.append(tempList)
                tempList=[]
            else:
                tempList.append(x)
                
    if len(tempList)>0:
        groups.append(tempList)
    else:
        pass
    #print(groups)
        
    return(groups)
```
