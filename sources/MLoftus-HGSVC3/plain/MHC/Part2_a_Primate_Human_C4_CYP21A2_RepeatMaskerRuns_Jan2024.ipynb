```python
tempList4=[]
with open('/home/mark/Desktop/HLA/C4_Master/Sample_Sequences_and_RepeatMasker/sample_geneRuns_RepeatMask/STK19_AllSamples_trimmed.out', 'r') as file:
    for line in file:
        tempList4.append(line.split())
file.close()
humanSTK19Gene = pd.DataFrame(data=tempList4)
```

```python
stk19Dict={}
for sample in set(humanSTK19Gene[5]):
    
    i=1
    stk19Dict[sample]={i:{'Start':0, 'End':0, 'Rows':[]}}
    
    tempDF = humanSTK19Gene[humanSTK19Gene[5]==sample].copy()
    tempDF[6]=tempDF[6].astype(int)
    tempDF[7]=tempDF[7].astype(int)
    tempDF.sort_values(by=[6], inplace=True)
    
    
    tempList=[]
    flag=0
    for row in tempDF.index:
        if flag == 0:
            tempList.append(row)
            currentExon = int(tempDF.at[row,10].split("EXON_")[1])
            flag+=1
        else:
            
            if int(tempDF.at[row,10].split("EXON_")[1])-1 == currentExon:
                tempList.append(row)
                currentExon = int(tempDF.at[row,10].split("EXON_")[1])
                
            else:
                
                tempDF2 = tempDF.loc[tempList].copy()
                stk19Dict[sample][i]['Start']=min(tempDF2[6])
                stk19Dict[sample][i]['End']=max(tempDF2[7])
                stk19Dict[sample][i]['Rows']=[[x for x in tempDF2.loc[row2]] for row2 in tempDF2.index]
                
                #Restart
                i+=1
                stk19Dict[sample][i]={'Start':0, 'End':0, 'Rows':[]}
                tempList=[]
                tempList.append(row)
                currentExon = int(tempDF.at[row,10].split("EXON_")[1])
                
    if len(tempList)>0:
        tempDF2 = tempDF.loc[tempList].copy()
        stk19Dict[sample][i]['Start']=min(tempDF2[6])
        stk19Dict[sample][i]['End']=max(tempDF2[7])
        stk19Dict[sample][i]['Rows']=[[x for x in tempDF2.loc[row2]] for row2 in tempDF2.index]
                
    else:
        continue
        
stk19Dict2={}
goodSTK19Rows=[]

for sample in stk19Dict.keys():
    
    stk19Dict2[sample]={}
    i=1
    
    for gene in stk19Dict[sample]:
        
        if len(stk19Dict[sample][gene]['Rows']) == 7:
            
            stk19Dict2[sample][i]={'Start':stk19Dict[sample][gene]['Start'], 'End':stk19Dict[sample][gene]['End'], 'Rows':stk19Dict[sample][gene]['Rows']}
            i+=1
            for exonHit in stk19Dict[sample][gene]['Rows']:
                goodSTK19Rows.append(exonHit)
        else:
            continue
```
