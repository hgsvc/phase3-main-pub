```python
import os
import pandas as pd
import pysam
```

```python
peter = pd.read_csv("/home/mark/Desktop/HLA/Globus/blast_hits_per_seq.tsv", sep='\t').set_index("fasta")
```

```python
directory='/home/mark/Desktop/HLA/Globus/original/'
sampleDictionary={}

for filename in os.listdir(directory):
    print(filename)
    totalLength=0
    currentlocation=1
    spanstart=[]
    sampleDictionary[filename]={'Total_Length':0, 'N-Sites':[], 'Start_Location':0, 'Peter_HLA_Hits':int(peter.at[str(filename)+".gz",'total_blast_hla_hits'])}
    
    with open(directory+str(filename), 'r') as file:
        with open('/home/mark/Desktop/HLA/Globus/cleaned/'+str(filename.split(".fasta")[0])+".cleaned.fasta", 'a+') as cleanfile:
            for line in file:
                if '>' in line[0]:
                    sampleDictionary[filename]['Start_Location']= int(line.split("_HLA_")[1].split("-")[1].split(":")[0])
                    print(line)
                    if 'haplotype' in str(line):
                        sample = line.split(">")[1].split(".")[0]
                        restOfName = line.split("sseq_haplotype")[1].split("-")[0]
                        cleanfile.write(">"+str(sample)+".hap"+str(restOfName)+"\n")

                    elif 'sseq_unassigned' in str(line):
                        sample = line.split(">")[1].split(".")[0]
                        restOfName = line.split("sseq_unassigned")[1]
                        cleanfile.write(">"+str(sample)+".unass"+'\n')
                    else:
                        print(line)
                else:
                    #continue
                    
                    for letter in [x for x in str(line.strip())]:
                        #print(letter)
                        if letter == 'N':
                            spanstart.append(currentlocation)
                            currentlocation+=1
                        else:
                            currentlocation+=1
                    
                    totalLength+=len(line.strip())
                    cleanfile.write(line)
                    
                    
            sampleDictionary[filename]['Total_Length']=totalLength
            sampleDictionary[filename]['N-Sites']=spanstart
        cleanfile.close()
    file.close()
```

```python
for key in sampleDictionary.keys():
    sampleDictionary[key]['Total_N_Sites']=len(sampleDictionary[key]['N-Sites'])
```

```python
for key in sampleDictionary.keys():
    sampleDictionary[key]['sequence_percent_N']=int(sampleDictionary[key]['Total_N_Sites'])/int(sampleDictionary[key]['Total_Length'])
```

```python
for key in sampleDictionary.keys():
    print(key, sampleDictionary[key]['sequence_percent_N'])
```

```python
import json
with open('/home/mark/Desktop/HLA/Globus/sample_hap_HLA_Sequence_Dictionary.json', 'w') as fp:
    json.dump(sampleDictionary, fp)
```
