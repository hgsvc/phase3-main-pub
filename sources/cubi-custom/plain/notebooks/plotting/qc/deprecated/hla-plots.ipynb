```python
%run "../00_project_config.ipynb"

import pathlib as pl
import pandas as pd
import collections as col
import hashlib as hl
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors

HLA_CHROM, HLA_START, HLA_END = load_ref_roi("t2t", "hla")

PLOT_OUTPUT_FOLDER = PROJECT_DATA_ROOT.joinpath("2023_plots/hla_roi")

def get_tigs():

    input_path = PROJECT_DATA_ROOT.joinpath("2023_ctgref_align/t2tv2")
    use_columns = [
        "query_name", "query_length", "query_start", "query_end",
        "target_name", "target_start", "target_end",
        "tp_align_type", "target_length"
    ]
    
    extract_table = []
    out_table = []
    for paf_file in sorted(input_path.glob("*norm-paf.tsv.gz")):
        sample = paf_file.name.split(".")[0]
        hap = paf_file.name.split(".")[2].split("-")[-1]
        paf = pd.read_csv(paf_file, sep="\t", header=0, comment="#", usecols=use_columns)
        paf = paf.loc[(paf["tp_align_type"] == 1) & (paf["target_name"] == HLA_CHROM), :].copy()
        paf = paf.loc[(paf["target_end"] > HLA_START) & (paf["target_start"] < HLA_END), :].copy()
        paf["ratio"] = (paf["query_length"] / paf["target_length"]).round(3)
                
        num_contigs = paf["query_name"].nunique()
        
        for query, aln in paf.groupby("query_name"):
            
            min_qstart = aln["query_start"].min()
            max_qend = aln["query_end"].max()
            min_tstart = aln["target_start"].min()
            max_tend = aln["target_end"].max()
            
            loc_length = round((max_qend - min_qstart)/1e6, 2)
            extract_table.append(
                (
                    query, min_qstart, max_qend, sample, loc_length,
                    min_tstart, max_tend, f"{query}:{min_qstart}-{max_qend}")
            )
            num_aln = aln.shape[0]
            ratio = aln["ratio"].values[0]
            out_table.append((sample, hap, num_contigs, query, num_aln, ratio))
    
    out_table = pd.DataFrame.from_records(
        out_table,
        columns=["sample", "hap", "num_contigs", "contig_name", "num_align", "size_ratio"]
    )
    
    extract_table = pd.DataFrame.from_records(
        extract_table, columns=[
            "contig", "qstart", "qend", "sample", "length",
            "ref_start", "ref_end", "region_spec"
        ]
    )
    
    extract_table.sort_values(["sample", "contig", "qstart"], inplace=True)
    extract_table.to_csv(
        PLOT_OUTPUT_FOLDER.joinpath("hla_coords_assemblies.tsv"),
        sep="\t", header=True, index=False
    )
    
    box_num_contigs = out_table.drop_duplicates(keep="first", inplace=False)["num_contigs"].values
    box_size_ratio = out_table["size_ratio"].values
    
    total = box_size_ratio.size
    frac_t2t = sum(box_size_ratio > 0.98)
        
    fig, ax = plt.subplots(figsize=(6,8))
    
    ax.boxplot(
        (box_num_contigs, box_size_ratio),
        positions=[1,2],
        meanline=True,
        labels=["Contigs (N)", "Size ratio"]
    )
    ax.text(1.75, 1.5, f"{total} total\n{frac_t2t} T2T\n(SR > 0.98)", fontsize=14)

    plt.savefig(PLOT_OUTPUT_FOLDER.joinpath("hla_contig_boxplot.pdf"), bbox_inches="tight")
    plt.savefig(
        PLOT_OUTPUT_FOLDER.joinpath("hla_contig_boxplot.png"),
        dpi=150, bbox_inches="tight"
    )
    
    return

#_ = get_tigs()


def plot_coverage():

    input_path = PROJECT_DATA_ROOT.joinpath("2023_ctgref_cov/t2tv2")
    
    hla_blocks_file = PROJECT_BASE.joinpath(
        "annotations/roi/suppl/t2tv2_hla_blocks_chr6.tsv"
    )
    hla_blocks = pd.read_csv(hla_blocks_file, sep="\t", header=0)
    
    coverages = []
    ylabels = []
    yticks = []
    
    xpos_labels = []
    
    region_begin = None
    region_end = None

    row_pos = 0
    for tsv_file in sorted(input_path.glob("*.tsv.gz")):
        df = pd.read_csv(tsv_file, sep="\t", header=[0,1,2], index_col=[0,1,2])
        df = df.loc[df.index.get_level_values("chrom") == HLA_CHROM, :].copy()
        df = df.droplevel("chrom", axis=0)
        df.reset_index(drop=False, inplace=True)
        df["start"] = df["start"].astype(int)
        offset = df["start"].min()
        df["end"] = df["end"].astype(int)
        df = df.loc[(df["start"] < HLA_END) & (df["end"] > HLA_START), :].copy()
        region_begin = df["start"].min()
        region_end = df["end"].max()
        sample = tsv_file.name.split(".")[0]

        for unit in ["hap1", "hap2", "unassigned"]:
            for mq in ["MQ60"]:
                values = df.loc[:, (f"asm-{unit}", mq, "ctg_align_cov")].values
                values = np.clip(values, 0, 3)  # everything > 2 is just "high"
                coverages.append(values)
                #labels.append(f"{sample} / {unit} / {mq}")
                if unit == "hap2":
                    ylabels.append(f"{sample}")
                    yticks.append(row_pos)
                row_pos += 1

    for i in range(region_begin, region_end, 10000):
        if i % 500000 == 0:
            xpos = int((i - region_begin)/10000)
            xpos_label = f"{i/1e6}"
            xpos_labels.append((xpos, xpos_label, "tick"))
    
    cov_hla = np.zeros_like(coverages[0], dtype=int)
    for row in hla_blocks.itertuples():
        block_start = (row.start - region_begin) // 10000
        block_end = (row.end - region_begin + 10000) // 10000
        assert block_start < block_end
        cov_hla[block_start:block_end] = -1
        hla_label_pos = int(np.floor(block_start + (block_end - block_start) / 2))
        xpos_labels.append((hla_label_pos, row.name.split("-")[-1], "label"))

    coverages.insert(0, cov_hla)
    
    coverages = np.vstack(coverages)

    colormap = colors.ListedColormap(["black", "white", "skyblue", "orange", "red"])

    fig, ax = plt.subplots(figsize=(25,25))

    ax.matshow(
        coverages,
        vmin=-1,
        vmax=3,
        cmap=colormap
    )

    xpos = []
    xlabels = []
    for idx, (pos, label, tick_type) in enumerate(sorted(xpos_labels)):
        if tick_type == "label":
            if idx == 0:
                xpos.append(pos)
                xlabels.append(label)
                continue
            dist_to_last = abs(pos - xpos[-1])
            if dist_to_last > 3:
                xpos.append(pos)
                xlabels.append(label)
            else:
                mrg = f"{xlabels[-1]}/{label}"
                xlabels[-1] = mrg
                
        else:
            dist_to_last = abs(pos - xpos[-1])
            if dist_to_last > 5:
                xpos.append(pos)
                xlabels.append(label)
            
    
    ax.set_xticks(xpos)
    ax.set_xticklabels(xlabels, fontsize=14, rotation=90)

    ax.set_xlabel(f"HLA / {HLA_CHROM} (Mbp)", fontsize=14)

    ax.set_yticks(yticks)
    ax.set_yticklabels(ylabels, fontsize=14)

    ax.set_ylabel("H1 / H2 / Un contigs (x-fold cov. in 10 kbp windows)", fontsize=14)
    
    plt.savefig(PLOT_OUTPUT_FOLDER.joinpath("hla_contig_coverage.pdf"), bbox_inches="tight")
    plt.savefig(
        PLOT_OUTPUT_FOLDER.joinpath("hla_contig_coverage.png"),
        dpi=150, bbox_inches="tight"
    )
    
    return

_ = plot_coverage()
```
